### Task 1.2: Key Assumptions

Assumption 1

Assumption: The most discriminative shapelet (local pattern separating classes) exists as a contiguous subsequence directly extractable from the training data.

Why the method needs it: Shapelet discovery relies on brute-force enumeration of training subsequences as candidates; without this, no motif can be found to compute gains and build the tree (core to Definitions 5 and 9).

Violation scenario: In datasets with class-specific patterns generated by transformations (e.g., rotated or scaled versions of a motif not present in training), like warped ECG signals where spikes are phase-shifted.

Paper reference: Section 2.1 formally defines a shapelet in Definition 9 as "a contiguous subsequence of a time series... that is maximally representative of a class," emphasizing extraction directly from training D for gain maximization. The motivating arrowheads example in Section 5.2 (Figure 3) illustrates this: two arrowhead classes differ by a local "barb" motif (subsequence of length ~10–20 points), explicitly extracted from training traces to achieve 80% accuracy, implying no transformative generation—violation would yield no high-gain candidates, as the method assumes raw extractability for brute-force validity.

Assumption 2

Assumption:Euclidean distance is an appropriate metric for subsequence similarity, assuming alignment without warping.

Why the method needs it: All distance calculations and gains use raw Euclidean (Definition 4), which underpins early abandon and pruning; warping would invalidate min-distance minima.

Violation scenario: Time-stretched series, e.g., variable-speed motion capture (Gun/NoGun stretched), where motifs align better with DTW than Euclidean.

Paper reference: Section 3.2 specifies SubsequenceDist in Definition 5 as the "minimum Euclidean distance" (Def 4: √∑(a_i - b_i)²) between a candidate and series subsequences, integrated into gain without alternatives. Table 6's early abandon pseudocode relies on this squared-error summation for partial pruning (e.g., partial_sum += (T[j] - S[k])²; if partial > min: abandon), assuming rigid alignment—paper notes in Sec 6 this limits to "univariate, non-warped" data, as warping (e.g., DTW) would require redefining minima, invalidating the O(1) abandon efficiency.

Assumption 3

Assumption: Classes are separable by one or few local discriminative patterns (shapelets), rather than global or distributed features.

Why the method needs it: The tree-building greedily selects single shapelets per node for splits; if separation requires ensemble of globals, gains would be low and tree shallow/poor.

Violation scenario: Datasets with global trends (e.g., overall amplitude differences across entire series, like coffee roast levels in Sec 5.5 but amplified).

Paper reference: Section 1 motivates shapelets as "local patterns that are representative of a class" contrasting full-series methods (e.g., 1-NN), claiming trees of "a handful of shapelets" suffice for motifs (e.g., "a single spike"). Figure 15's shields example (Sec 5.5) relies on local heraldry motifs for 92% accuracy, with global baselines failing—implicitly assuming few locals drive separation, as greedy selection (Sec 4, Table 8) yields shallow trees (3–5 nodes); globals would dilute gains (Def 7 <0.1), leading to impure splits and poor scaling.
